```python
# Iris Flower Classification
# This project trains a machine learning model to classify Iris flowers
# based on their sepal and petal measurements.
# The model is trained on known flower data and tested on unseen samples
# to evaluate its prediction accuracy.


from sklearn.datasets import load_iris

iris = load_iris()
print(iris.keys())

x = iris.data
y = iris.target

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_pred))
```
